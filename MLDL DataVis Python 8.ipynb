## In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
%matplotlib inline
```

```python
dataset = datasets.load_iris()
```

```python
dataset
```

```python
x = PCA(n_components=3).fit_transform(dataset.data)
y = dataset.target
```

```python
x
```

```python
y
```

```python
figure = plt.figure(1, figsize=(8,8))
axis = figure.add_subplot(projection='3d')
axis.scatter(x[:,0], x[:,1], x[:,2], c=y, cmap=plt.cm.rainbow)
axis.set_title("PCA 3D")
axis.set_xlabel("PCA X")
axis.set_ylabel("PCA Y")
axis.set_zlabel("PCA Z")
```
